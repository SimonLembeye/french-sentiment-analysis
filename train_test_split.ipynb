```python
import csv
import numpy as np
```

```python
raw_data_path = "/home/simon/Bureau/French-Sentiment-Analysis-Dataset/tweets.csv"
data_path = "data"
```

```python
f = open(raw_data_path, "r")
lines = f.readlines()
```

```python
n = len(lines)
n
```

```python
perm = np.random.permutation(range(1, n))
dev_indices = perm[:10000]
test_indices = perm[10000:20000]
```

```python
count = 0
train = []
dev = []
test = []

for i in range(1, n):
    arr = lines[i].split(",")
    if arr[0] == "0":
        label = 0
    elif arr[0] == "4":
        label = 1
    else:
        print("No Label !")
    
    sentence = ",".join(arr[1:])
    
    if label:
        line = str(label) + "," + sentence
        if i in dev:
            dev.append(line)
        elif i in test:
            test.append(line)
        else:
            train.append(line)
```

```python
with open(data_path + "/train.csv", "w") as csv_file:
    for line in train:
        csv_file.write(line)
```

```python
with open(data_path + "/dev.csv", "w") as csv_file:
    for line in dev:
        csv_file.write(line)
```

```python
with open(data_path + "/test.csv", "w") as csv_file:
    for line in dev:
        csv_file.write(line)
```
